```python
AgeGroup={}
length=df1.shape[0]
AgeGroup['25-30']=round(df1[(df1['Age']>=25)&(df1['Age']<30)].shape[0]*100/length,2)
AgeGroup['30-35']=round(df1[(df1['Age']>=30)&(df1['Age']<35)].shape[0]*100/length,2)
AgeGroup['35-40']=round(df1[(df1['Age']>=35)&(df1['Age']<40)].shape[0]*100/length,2)
AgeGroup['40-45']=round(df1[(df1['Age']>=40)&(df1['Age']<45)].shape[0]*100/length,2)
AgeGroup['45-50']=round(df1[(df1['Age']>=45)&(df1['Age']<50)].shape[0]*100/length,2)
AgeGroup['50-55']=round(df1[(df1['Age']>=50)&(df1['Age']<55)].shape[0]*100/length,2)
AgeGroup['55+']=round(df1[df1['Age']>=55].shape[0]/length,2)
```
